### dependencies

```python
import numpy as np
import pandas as pd
import os
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import pickle
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout, Activation, BatchNormalization
import gc
import warnings
warnings.filterwarnings('ignore')
```

### creating the dataset

```python
path = '../input/amazonsatelliteimages/train-jpg/train-jpg'

for img in os.listdir(path):
    img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
    plt.imshow(img_arr)
    # printing one image
    break
```

```python
gc.collect()
IMG_SIZE = 50

train_data = []

def create_train_data():

    for img in os.listdir(path):
        
        try:
            img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            new_img_arr = cv2.resize(img_arr, (IMG_SIZE, IMG_SIZE))
            train_data.append(new_img_arr)
        except Exception as e:
            print(e)
        
create_train_data()
```

```python
labels = pd.read_csv('../input/planet-understanding-the-amazon-from-space/train_v2.csv/train_v2.csv')

X = np.array(train_data).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
y = pd.DataFrame(labels['tags'])

X.shape
```

```python
X = X[20000:,:]
y = y[20000:]
```

### encoding output variable

```python
encode = LabelEncoder().fit(y)
encoded_y = encode.transform(y)
y = np_utils.to_categorical(encoded_y)

pickle.dump(X, open('./X.pkl','wb'))
pickle.dump(y, open('./y.pkl','wb'))

y
```

### creating the model

```python
train_X = pickle.load(open('./X.pkl', 'rb'))
train_y = pickle.load(open('./y.pkl', 'rb'))

# normalizing features
train_X = tf.keras.utils.normalize(train_X, axis=1)

n_epochs = 10
b_size = 2000

model = Sequential()
# ADDING LAYERS

# input layer
model.add(BatchNormalization(input_shape=train_X.shape[1:]))
model.add(Conv2D(64, (3, 3), padding='same', kernel_initializer='he_uniform'))
model.add(Conv2D(64, (3, 3), padding='same', kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, (3, 3), padding='same', kernel_initializer='he_uniform'))
model.add(Conv2D(128, (3, 3), padding='same', kernel_initializer='he_uniform'))
model.add(Conv2D(128, (3, 3), padding='same', kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.5))

model.add(Conv2D(256, (3, 3), padding='same'))
model.add(Conv2D(256, (3, 3), padding='same'))
model.add(Conv2D(256, (3, 3), padding='same'))
model.add(Conv2D(256, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.5))

model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Dropout(0.5))

model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Conv2D(512, (4, 4), padding='same'))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.5))

# output neuron
model.add(Flatten())
model.add(Dense(4096,activation="softmax"))
model.add(Dense(4096,activation="softmax"))

model.add(Dense(347))
model.add(Activation('softmax'))

opt = optimizers.Adam(lr=1e-3)
model.compile(optimizer=opt,
              loss='kullback_leibler_divergence',
              metrics=['accuracy'])
```

```python
gc.collect()
model.summary()
```

### training the model

```python
gc.collect()
```

```python
train = model.fit(train_X, train_y, epochs=n_epochs, batch_size=b_size, validation_split=0.2, shuffle=True)
train
plt.figure()
plt.plot(train.history['loss'], 'blue')
plt.plot(train.history['val_loss'], 'red')
plt.legend(['Training Loss', 'Validation Loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

```python
plt.figure()
plt.plot(train.history['accuracy'], 'blue')
plt.plot(train.history['val_accuracy'], 'red')
plt.legend(['Training Accuracy', 'Validation Accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```

### calculating fbeta score

```python
test_y = train_y
```

```python
from sklearn.metrics import fbeta_score
fbeta_score(test_y, y_pred, average='macro', beta=0.5)

def fbeta(test_y, y_pred, beta=2, threshold=0.2, epsilon=1e-7):
    # threshold is set to 0.2 to maximize recall since f2 score is recall biased
    # epsilon is set to 1e-7 to avoide Nan values due to zero division
    
    beta_squared = float(beta)**2
    
    test_y = tf.cast(test_y, tf.float32) # casts test_y as a float
    # convert y_pred to bool, then to float
    y_pred = tf.cast(tf.greater(tf.cast(y_pred, tf.float32), tf.constant(threshold)), tf.float32) 
    
    tp = tf.reduce_sum(tf.cast(tf.equal((2.0*test_y + y_pred), tf.constant(3.0)), tf.float32), axis=1) 
    fp = tf.reduce_sum(tf.cast(tf.equal((2.0*test_y + y_pred), tf.constant(1.0)), tf.float32), axis=1)
    fn = tf.reduce_sum(tf.cast(tf.equal((2.0*test_y + y_pred), tf.constant(2.0)), tf.float32), axis=1)

    precision = tp / (tp+fp)
    recall = tp / (tp+fn)
    fb = (beta_squared+1) * precision * recall / (precision*beta_squared + recall + epsilon)
    
    print(fb)
    return fb

fbeta(test_y, y_pred)
```
